```python
import requests

base_url = 'https://collectionapi.metmuseum.org'
search_path = '/public/collection/v1/search'
query ='?isHighlight=true&q=*'
r = requests.get(f'{base_url}{search_path}{query}')
#r is response
parsed = r.json()
#parse r
ids = parsed['objectIDs']
```

```python
parsed
```

```python
import requests
import time

listobjects = []
base_url = 'https://collectionapi.metmuseum.org'
path = '/public/collection/v1/objects/'
t = 2

for item in ids:
    url = f'{base_url}{path}{item}'

    response = requests.get(url)

    if response.status_code == 200:
        parsed = response.json()
        listobjects.append(parsed)
    else:
        print(f"Error fetching object {item}: Status code {response.status_code}")

    time.sleep(t)
```

```python
genders = []
titles = []
names = []
begins = []
ends = []
departments = []
periods = []
cultures = []
countries = []
for i in listobjects:
    title = i.get('title')
    titles.append(title)
    gender = i.get('artistGender')
    genders.append(gender)
    name = i.get('artistDisplayName')
    names.append(name)
    begin = i.get('objectBeginDate')
    begins.append(begin)
    end = i.get('objectEndDate')
    ends.append(end)
    department = i.get('department')
    departments.append(department)
    period = i.get('period')
    periods.append(period)
    culture = i.get('culture')
    cultures.append(culture)
    country = i.get('country')
    countries.append(country)
```

```python
import pandas as pd

df = pd.DataFrame({
    'title': titles,
    'name': names,
    'gender' : genders,
    'begin': begins,
    'end': ends,
    'department': departments,
    'culture':cultures,
    'country':countries
})
```
